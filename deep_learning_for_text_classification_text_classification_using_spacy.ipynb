```python
import pandas as pd
train_data=pd.read_csv("/content/nlp_train_deeplearning.csv")
test_data=pd.read_csv("/content/nlp_test_deeplearning.csv")
train_data.head()
```

```python
!pip install spacy
!python -m spacy download en_core_web_lg
import spacy
model=spacy.load("en_core_web_lg") #what it means is that it is loading a pre trained statistical model for the english language, the en in it is what tells it that it is off the english language.
```

```python
#making a new column in the dataset which would store the vector of the of each row
train_data['vector']=train_data['text'].apply(lambda text: model(text).vector)
test_data['vector']=test_data['text'].apply(lambda text: model(text).vector)
```

```python
train_data.head()
```

```python
x_train=train_data.vector.values
x_test=test_data.vector.values
y_train=train_data.target
```

```python
y_train
```

```python
import numpy as np
x_train=np.stack(x_train)
x_test=np.stack(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
clf=MultinomialNB()
clf.fit(x_train,y_train)
#Why did I use minmaxscaler, because multinomialNB does not accept negative values
#what is multinomialnb :multinomial nb is a model under naive_bayes which calculates the probability distribution of the text data, it uses bayes theorem
```

```python
y_predict=clf.predict(x_test)
```

```python
accuracy = clf.score(x_train, y_train)
print(accuracy)
```
